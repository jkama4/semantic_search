```
def generate_responses() -> List[str]:
    responses: List[str] = []

    for query, _ in QUERY_AND_ANSWERS:
        print(f"Generating response for: {query}")

        response = requests.post(
            url=config.LLM_ENDPOINT,
            json={
                "history": [{"role": "user", "content": query}],
                "inject_info": True,
                "rerank": False,
            },
        )
        response.raise_for_status()

        print(response)

        responses.append(response.json()["response"])

    return responses
```
